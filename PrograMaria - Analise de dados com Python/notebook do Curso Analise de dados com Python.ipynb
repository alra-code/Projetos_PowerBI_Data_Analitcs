#Intervalo de confiança e distribuição amostral

```python
salarios = dados['SALARIO']
salarios
```

```python
#importação
import numpy as np
```

```python
#media
media_amostral = np.mean(salarios)
media_amostral
```

```python
#desvio
desvio_amostral = np.std(salarios)
desvio_amostral
```

```python
#tamanho da amostra
tamanho_amostral = len(salarios)
tamanho_amostral
```

```python
#erro padrão
from scipy import stats
erro_padrao = stats.sem(salarios)
erro_padrao
```

```python
#intervalo de confiança
intervalo_confianca = stats.t.interval(alpha=0.95, df=tamanho_amostral-1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
```

```python
#Criando uma nova coluna
def determinar_etnia(cor_raca_etnia):
  if cor_raca_etnia == 'Branca':
    return 'Branca'
  elif cor_raca_etnia == 'Outra':
    return 'Outra'
  else:
    return 'Não Branca'
```

```python
#Featuring engineering do incio do exercício
def preencher_nivel(gestor,nivel):
  if pd.isnull(gestor):
    return nivel
  else:
    return gestor
  if gestor == 1:
    return "Pessoa Gestora"
  else:
    return nivel
```

#Criando nova coluna

```python
def determinar_geracao(idade):
  if 39<idade <= 58:
    return 'Geração x'
  elif 29<idade <= 39:
    return 'Geração Millenial'
  elif 13<idade <= 29:
    return 'Geração Z'
  else :
    return 'Geração Outra Geração'
```

#Correlação

```python
correlacao_continua = dados['IDADE'].corr(dados['SALARIO'])
correlacao_continua
```

```python
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente (coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(dados[coluna1], dados[coluna2]))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1
  cramer = np.sqrt(chic2/(soma*mini))
  return cramer
```

```python
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```
